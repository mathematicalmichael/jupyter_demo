# <center> Example Plots with Octave

```octave
x = -10:0.1:10;
plot (x, sin (x));
```

```octave
hist (randn (10000, 1), 30);
```

```octave
h = bar (rand (5, 10));
set (h(1), "basevalue", 0.5);
```

```octave
x = 1:10;
y = 2*x;
stem (x, y, "r");
```

```octave
theta = 0:0.2:6;
stem3 (cos (theta), sin (theta), theta);
```

```octave
[th, r] = rose ([2*randn(1e5,1), pi + 2*randn(1e5,1)]);
polar (th, r);
```

```octave
polar (0:0.1:10*pi, 0:0.1:10*pi);
```

```octave
x = 0:2;
y = x;
z = x' * y;
contour (x, y, z, 2:3)
```

```octave
contour3 (peaks (19));
colormap cool;
hold on;
surf (peaks (19), "facecolor", "none", "edgecolor", "black");
```

```octave
x = 0:0.1:10;
y = sin (x);
yp =  0.1 .* randn (size (x));
ym = -0.1 .* randn (size (x));
errorbar (x, sin (x), ym, yp);
```

```octave
[x, y] = meshgrid (1:2:20);
h = quiver (x, y, sin (2*pi*x/10), sin (2*pi*y/10));
set (h, "maxheadsize", 0.33);
```

```octave
[x, y, z] = peaks (25);
surf (x, y, z);
hold on;
[u, v, w] = surfnorm (x, y, z / 10);
h = quiver3 (x, y, z, u, v, w);
set (h, "maxheadsize", 0.33);
```

```octave
phi = [0 : 15 : 360] * pi/180;
feather (sin (phi), cos (phi));
```

```octave
t = linspace (0, 2*pi, 100)';
y = [sin(t).^2, cos(t).^2];
area (t, y);
legend ("sin^2", "cos^2", "location", "NorthEastOutside");
```

```octave
vertices = [0 0
            1 0
            1 1
            0 1];
fill (vertices(:,1), vertices(:,2), "r");
axis ([-0.5 1.5, -0.5 1.5])
axis equal
```
